## Beatiful Soup

[Beautiful Soup Docs](https://beautiful-soup-4.readthedocs.io/en/latest/#modifying-the-tree)

```python
from bs4 import BeautifulSoup
import requests

page = requests.get("https://es.wikipedia.org/wiki/Python")

soup = BeautifulSoup(page.content, "html.parser")

print(soup.prettify())
```

```python
# Sin anidación
list(soup.children)
```

```python
enlaces = soup.findAll("div", {"class": "listaref"})
enlaces
```

```python
soup.find_all("li")
```

```python
for d in soup.find("div", attrs={"class": "listaref"}).descendants:
    if d.name == "span" and d.get("class", "") == ["reference-text"]:
        print(d.text)
```

```python
data = soup.find("div", attrs={"class": "listaref"})
for div in data:
    links = div.findAll("a")
    for a in links:
        if a["href"].startswith("http"):
            print(a["href"])
```

```python
soup.find("body")
```

```python
soup.find("body").find_all("p")
```

```python
parrafos = soup.find("body").find_all("p")
```

```python
import re

def limpiar(html):
    exp = re.compile('<.*?>')
    texto = re.sub(exp, '', html)
    return texto
```

```python
for parrafo in parrafos:
    print(limpiar(str(parrafo)))
```
